```python
import matplotlib.image as mpimg 
import matplotlib.pyplot as plt 
import numpy as np
import imgaug.augmenters as iaa
import cv2
import statistics
```

```python
P = cv2.imread('P.png',0)
_, P = cv2.threshold(P, 128, 255, cv2.THRESH_BINARY_INV)
plt.imshow(P) 
```

```python
R = cv2.imread('R.png',0)
_, R = cv2.threshold(R, 128, 255, cv2.THRESH_BINARY_INV)
plt.imshow(R)
```

```python
W = cv2.imread('W.png',0)
_, W = cv2.threshold(W, 128, 255, cv2.THRESH_BINARY_INV)
plt.imshow(W)
```

```python
seq = iaa.Sequential([
    iaa.Sometimes(
        0.5,
        iaa.GaussianBlur(sigma=(0, 0.5))
    ),
    iaa.Affine(
        scale={"x": (0.75, 1.25), "y": (0.75, 1.25)},
        translate_percent={"x": (-0.2, 0.2), "y": (-0.2, 0.2)},
        rotate=(-45, 45),
    )
], random_order=True)
```

```python
P_cs = [seq(images=np.array(P)) for _ in range(10)]
W_cs = [seq(images=np.array(W)) for _ in range(10)]
R_cs = [seq(images=np.array(R)) for _ in range(10)]
```

```python
def get_hu_moments(imgs, imgs_names, hu_nr):
    results = {}
    for imgs_nr in range(len(imgs)):
        results[imgs_names[imgs_nr]] = {}
        for img_nr in range(len(imgs[imgs_nr])):
            moments = cv2.moments(imgs[imgs_nr][img_nr])
            hu_moment = cv2.HuMoments(moments)[hu_nr][0]
            results[imgs_names[imgs_nr]][img_nr] = hu_moment
        results[imgs_names[imgs_nr]]["Var"] = statistics.variance(results[imgs_names[imgs_nr]].values())
    results["VAR"] = {}
    for i in range(7):
        results["VAR"][i] = statistics.variance([results[img][i] for img in imgs_names]) 
    return results
```

```python
import pandas as pd
import math
results = [get_hu_moments([P_cs, W_cs, R_cs], ['P','W','R'], hu_nr) for hu_nr in range(7)]
dfs = [pd.DataFrame(data=result) for result in results]
```
