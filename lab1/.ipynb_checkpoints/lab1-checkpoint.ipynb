```python
import matplotlib.image as mpimg 
import matplotlib.pyplot as plt 
import numpy as np
import imgaug.augmenters as iaa
import cv2
```

```python
P = cv2.imread('P.png',0)
plt.imshow(P) 
```

```python
R = cv2.imread('R.png',0)
plt.imshow(R)
```

```python
W = cv2.imread('W.png',0)
plt.imshow(W)
```

```python
seq = iaa.Sequential([
    iaa.Sometimes(
        0.5,
        iaa.GaussianBlur(sigma=(0, 0.5))
    ),
    iaa.Affine(
        scale={"x": (0.8, 1.2), "y": (0.8, 1.2)},
        translate_percent={"x": (-0.2, 0.2), "y": (-0.2, 0.2)},
        rotate=(-25, 25),
        shear=(-8, 8)
    )
], random_order=True)
```

```python
_, P_t = cv2.threshold(P, 128, 255, cv2.THRESH_BINARY_INV)
_, W_t = cv2.threshold(W, 128, 255, cv2.THRESH_BINARY_INV)
_, R_t = cv2.threshold(R, 128, 255, cv2.THRESH_BINARY_INV)

P_cs = [seq(images=np.array(P_t)) for _ in range(10)]
W_cs = [seq(images=np.array(W_t)) for _ in range(10)]
R_cs = [seq(images=np.array(R_t)) for _ in range(10)]
```

```python
plt.imshow(W_cs[0])
```

```python
results = {}

def get_hu_moments(imgs):
    results = {}
    for i in range(len(imgs)):
        results[i] = {}
        for j in range(len(imgs[i])):
            moments = cv2.moments(imgs[i][j])
            results[i][j] = hu_moments = cv2.HuMoments(moments)
            
    return results
```

```python
print(get_hu_moments([P_cs, W_cs, R_cs]))
```
